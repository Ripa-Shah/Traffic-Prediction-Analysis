```python
# Testing simlation of generating random points 
from __future__ import division
import numpy as np
import datetime
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
import random
import pandas as pd

def create_random_point(x0,y0,distance):
    """
            Utility method for simulation of the points
    """   
    r = distance/ 111300
    u = np.random.uniform(0,1)
    v = np.random.uniform(0,1)
    w = r * np.sqrt(u)
    t = 2 * np.pi * v
    x = w * np.cos(t)
    x1 = x / np.cos(y0)
    y = w * np.sin(t)
    return (x0+x1, y0 +y)

fig = plt.figure()
ax = host_subplot(111, axes_class=AA.Axes)
ax.set_autoscale_on(True)

latitude1,longitude1 = 33.787914 ,-117.853104

ax.plot(latitude1,longitude1,'ro')
L=[]
for i in range(1,100):
    x,y = create_random_point(latitude1,longitude1 ,500 )
    ax.plot(x,y,'bo')
    #df = pd.DataFrame(columns=['Number','TrafficRange','Latitude','Longitude','TimeSeries'])
    d={'Number':i,'TrafficRange':random.randint(0,100),'Latitude': x, 'Longitude':y}
    #print(df.head())
    df.to_csv(r"C:\Users\19498\Downloads\out.csv",index=False,header=True)
    L.append(d)
    df=pd.DataFrame(L)
    #ans=str(i)+'   '+str(random.randint(0,50))+'   '+str(x)+'   '+str(y)
    #print(ans)
    #plt.show()
```
